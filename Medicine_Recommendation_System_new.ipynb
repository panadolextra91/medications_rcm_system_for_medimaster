# Training top models

```python
%pip install numpy

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np


# Create a dictionary to store models
models = {
    'SVC': SVC(kernel='linear'),
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'GradientBoosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'KNeighbors': KNeighborsClassifier(n_neighbors=5),
    'MultinomialNB': MultinomialNB()
}

# Loop through the models, train, test, and print results
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Test the model
    predictions = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, predictions)
    print(f"{model_name} Accuracy: {accuracy}")

    # Calculate confusion matrix
    cm = confusion_matrix(y_test, predictions)
    print(f"{model_name} Confusion Matrix:")
    print(np.array2string(cm, separator=', '))

    print("\n" + "="*40 + "\n")
```

```python
import pandas as pd
import ast

# 1) Load your original disease→med list
df = pd.read_csv("/content/drive/MyDrive/prethesis/medications_filtered.csv")

# 2) Define the “allowed” medications (i.e. your 16 category names)
allowed = [
    'Analgesics','Antibiotics','Antihistamines','Cough suppressants', "Cold compresses",
    'Digestive Health','Heart & Blood Pressure','Diabetes',
    'Antiseptics','Vitamins & Supplements','First Aid',
    'Antifungal Cream','Antacids','Antiviral drugs','Antimalarial drugs',
    'Antispasmodics','Neuro', 'Urinary analgesics', 'Topical treatments', 'Pain relievers', 'Antipyretics', 'Probiotics'
]

# 3) Parse & filter
def filter_to_allowed(med_str):
    # turn the string "['A','B',...]" into a real list
    meds = ast.literal_eval(med_str)
    # only keep those in your allowed list
    return [m for m in meds if m in allowed]

df['Medication'] = df['Medication'].apply(filter_to_allowed)

df.to_csv(
    "/content/drive/MyDrive/prethesis/medications_categories_filtered.csv",
    index=False
)
```

```python
import pandas as pd
import ast

# 1. Load the filtered CSV
df = pd.read_csv("/content/drive/MyDrive/prethesis/medications_categories_filtered.csv")

# 2. Define a function to transform each medication list
def transform_med_list(med_str):
    meds = ast.literal_eval(med_str)            # parse "['A','B',...]" → list
    new = []
    for m in meds:
        if m == 'Pain relievers':
            new_m = 'Analgesics'
        elif m.endswith(' drugs'):
            new_m = m[:-6]                     # drop the " drugs" suffix
        elif m in ('Cold compresses', 'Cough suppressants'):
            new_m = 'Cough & Cold'
        elif m == 'Probiotics':
            new_m = 'Digestive Health'
        elif m == 'Antifungal Cream'    :
            new_m = 'Antifungal'
        else:
            new_m = m
        new.append(new_m)
    # remove duplicates while preserving order
    seen = set(); result = []
    for x in new:
        if x not in seen:
            seen.add(x)
            result.append(x)
    return result

# 3. Apply the transformation
df['Medication'] = df['Medication'].apply(transform_med_list)

# 4. (Optional) If you want to store them back as literal lists in CSV:
#df['Medication'] = df['Medication'].apply(lambda lst: str(lst))

# 5. Save the updated file
df.to_csv("/content/drive/MyDrive/prethesis/medications_categories_updated.csv", index=False)
df
```
